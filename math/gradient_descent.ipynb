```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.rand(100, 1)
noise = np.random.normal(size=(100, 1))
y = 12 * X + 3 + noise

plt.scatter(X, y)

ones = np.ones(X.shape)
X = np.hstack([X, ones])
X.shape
```

```python
# the exact descent update rule depends on the loss function used
# for the simplest case, linear regression
# one natural choice of loss function is the mean squared error

epoch = 10000
lr = 0.0001

theta = np.random.rand(2, 1)
mse_loss_hist = []
for i in range(epoch):
    pred = X @ theta # (100, 1)
    mse_loss = np.sum((pred - y) ** 2) / len(y)
    theta = theta - lr * X.T @ (pred - y)
    mse_loss_hist.append(mse_loss)

print(theta)

plt.plot(mse_loss_hist)
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("Loss Curve of Linear Regression with Gradient Descent")
```

```python
# visualize the line with the scatter plot
_x = np.linspace(0, 1, 100).reshape(-1, 1)
_x = np.hstack([_x, ones])

y_line = _x @ theta
y_pred = X @ theta
mse = np.sum((y_pred - y) ** 2) / len(y)
mae = np.sum(np.abs(y_pred - y)) / len(y)

print(f"MSE = {mse:.5f}")
print(f"MAE = {mae:.5f}")

plt.plot(_x[:, 0], y_line, 'r', label="best fit line")
plt.scatter(X[:, 0], y, label="data")
plt.title("Linear Regression Results")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
