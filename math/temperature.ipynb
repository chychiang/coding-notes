# Applying temperature scaling to the output logit

```python
import numpy as np
import matplotlib.pyplot as plt


logits = np.random.normal(size=(50))


data_list = [
    np.exp(logits) / np.sum(np.exp(logits)),
    np.exp(logits / 2) / np.sum(np.exp(logits / 2)),
    np.exp(logits / 0.5) / np.sum(np.exp(logits / 0.5)),
]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))
for ax, data in zip((ax1, ax2, ax3), data_list):
    ax.bar(np.arange(len(logits)), data)
```
